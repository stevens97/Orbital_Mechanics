## Theoretical Background

The orbital path of the comet can be approximated as an ellips on a two-dimensional plane (x,y) to simplify calculations. This can be done, since the calculation of the best suited time-step corresponding to each initial velocity doesn't require higher-dimensional calculations.

**According to Newton's universal law of gravitation**, the force that would be applied on the comet would be:

$$F_{x} = \frac{-G.M.m.r_{x}}{r^{3}} \ \ [N]$$

and

$$F_{y} = \frac{-G.M.m.r_{y}}{r^{3}} \ \ [N]$$

Where:

$F_{x}$: The force in the x-direction. $[N]$

$F_{y}$: The force in the y-direction. $[N]$

$G$: Newton's gravitational constant. $[s^{2}.m^{-1}.kg^{-1}]$

$M$: The mass of the heavier body (in this case: The Sun). $[kg]$

$m$: The mass of the lighter body (in this case: Halley's comet). $[kg]$

$r_{x}$: The x-component of the radial distance of the orbit. $[m]$

$r_{y}$: The y-component of the radial distance of the orbit. $[m]$

In astronomical units:
$GM \approx 4 \pi^{2} AU^{3}/yr^{2}$

**According to Newton's 2nd law of motion:**

$$F = m.a$$

Where:

$F$: The force applied to the object. $[N]$

$m$: The mass of the object. $[kg]$

$a$: The acceleration of the object. $[m.s^{-2}]$

Therefore, the **acceleration of the comet due to gravity** is:

$$a_{x} = \frac{-G.M.r_{x}}{r^{3}} = \frac{-G.M.r_{x}}{
(\sqrt{r_{x}^{2}+r_{y}^{2}})^{3}}$$

$$a_{y} = \frac{-G.M.r_{y}}{r^{3}} = \frac{-G.M.r_{y}}{
(\sqrt{r_{x}^{2}+r_{y}^{2}})^{3}}$$

Where:

$a_{x}$: The x-component of the acceleration of the comet. $[m.s^{-2}]$

$a_{y}$: The y-component of the acceleration of the comet. $[m.s^{-2}]$

According to **Kepler's laws**:

The velocity of the comet in orbit, in terms of the orbital radius r is:

$$v = \sqrt{GM(\frac{2}{r} - \frac{1}{a})}$$

And the Perihelion (closest approach) $q$ and Aphelion (Furthest approach) $Q$ can be calculated in terms of the eccentricity of the orbit: 

$$q = (1-e)a$$

$$Q = (1+e)a$$

**For Halley's comet**: 

$q = 0.587$

*Which yields...*

*With corresponding Perihelion and Aphelion velocities of:*

# Method

To answer Question 3.6, the **Euler-Cromer** method needs to utilised and to answer Question 3.11, the **4th-Order Runge-Kutta** method needs to be utilised.

Therefore a numerical scheme to approximate the values of the velocity $(v)$ and the radial position $(r)$ of the comet need to obtained for each method.

## The Euler-Cromer Method

The trajectory of the comet can be described by the following set of differential equations:

$$\frac{dr}{dt} = v$$

$$\frac{dv}{dt} = a(r)$$

Where:

$r:$ The radial position of the comet. $[AU]$

$v:$ The velocity of the comet. $[AU / year]$

$a:$ The acceleration of the comet. $[AU / year^{2}]$

Applying the Euler-Cromer method, **the following numerical scheme is obtained** to approximate the values of v and r, using a time-step of $\tau$:

$$a_{x_{i}}(r_{x_{i}}, r_{y_{i}}) = \frac{-G.M.r_{x_{i}}}{(\sqrt{r_{x_{i}}^{2}+r_{y_{i}}^{2}})^{3}}$$

$$a_{x_{i}}(r_{x_{i}}, r_{y_{i}})  = \frac{-G.M.r_{y_{i}}}{(\sqrt{r_{x_{i}}^{2}+r_{y_{i}}^{2}})^{3}}$$

$$v_{x_{i+1}} \approx v_{x_i} + \tau.a(r_{x_i}, r_{y_i})$$

$$v_{y_{i+1}} \approx v_{y_i} + \tau.a(r_{x_i}, r_{y_i})$$

$$r_{x_{i+1}} \approx r_{x_{i}} + \tau.v_{x_{i+1}}$$

$$r_{y_{i+1}} \approx r_{y_{i}} + \tau.v_{y_{i+1}}$$

Where:

$a_{x_{i}} = a_{x}(t_{i})$

$a_{y_{i}} = a_{y}(t_{i})$

$v_{x_{i+1}} = v_{x}(t_{i}+\tau)$

$v_{y_{i+1}} = v_{y}(t_{i}+\tau)$

$v_{x_{i}} = v_{x}(t_{i})$

$v_{y_{i}} = v_{y}(t_{i})$

$r_{x_{i+1}} = r_{x}(t_{i}+\tau)$

$r_{y_{i+1}} = r_{y}(t_{i}+\tau)$

$r_{x_{i}} = r_{x}(t_{i})$

$r_{y_{i}} = r_{y}(t_{i})$

$t_{i} = i.\tau$

## The 4th Order Runge-Kutta Method:

The same set of differential equations apply.

In this case however, **the following numerical scheme is obtained** to approximate the values of v and r, using a time-step of $\tau$.

$$v_{x_{i+1}} \approx v_{x_{i}} + \frac{\tau}{6}[ K_{1_{v_x}} + 2.K_{2_{v_x}} + 2.K_{3_{v_x}} + K_{4_{v_x}} ]$$

$$v_{y_{i+1}} \approx v_{y_{i}} + \frac{\tau}{6}[ K_{1_{v_y}} + 2.K_{2_{v_y}} + 2.K_{3_{v_y}} + K_{4_{v_y}} ]$$

$$r_{x_{i+1}} \approx r_{x_{i}} + \frac{\tau}{6}[ K_{1_{r_x}} + 2.K_{2_{r_x}} + 2.K_{3_{r_x}} + K_{4_{r_x}} ]$$

$$r_{y_{i+1}} \approx r_{y_{i}} + \frac{\tau}{6}[ K_{1_{r_y}} + 2.K_{2_{r_y}} + 2.K_{3_{r_y}} + K_{4_{r_y}} ]$$

**Where:**

$$K_{1_{v_x}} = a_{x}(r_{x_{i}}, r_{y_{i}}) = \frac{-G.M.r_{x}}{
(\sqrt{r_{x}^{2}+r_{y}^{2}})^{3}} $$

$$K_{1_{v_y}} = a_{y}(r_{x_{i}}, r_{y_{i}}) = \frac{-G.M.r_{y}}{
(\sqrt{r_{x}^{2}+r_{y}^{2}})^{3}} $$

$$K_{2_{v_x}} = a_{x}(r_{x_{i}} + \frac{\tau}{2}.K_{1_{r_x} }, r_{y_{i}} + \frac{\tau}{2}.K_{1_{r_y}})$$

$$K_{2_{v_y}} = a_{y}(r_{x_{i}} + \frac{\tau}{2}.K_{1_{r_x} }, r_{y_{i}} + \frac{\tau}{2}.K_{1_{r_y}})$$

$$K_{3_{v_x}} = a_{x}(r_{x_{i}} + \frac{\tau}{2}.K_{2_{r_x} }, r_{y_{i}} + \frac{\tau}{2}.K_{2_{r_y}})$$

$$K_{3_{v_y}} = a_{y}(r_{x_{i}} + \frac{\tau}{2}.K_{2_{r_x} }, r_{y_{i}} + \frac{\tau}{2}.K_{2_{r_y}})$$

$$K_{4_{v_x}} = a_{x}(r_{x_{i}} + \frac{\tau}{2}.K_{2_{r_x} }, r_{y_{i}} + \frac{\tau}{2}.K_{2_{r_y}})$$

$$K_{4_{v_y}} = a_{y}(r_{x_{i}} + \tau.K_{3_{r_x} }, r_{y_{i}} + \tau.K_{3_{r_y}})$$

**And**


$$K_{1_{r_x}} = v_{x_{i}}$$

$$K_{1_{r_y}} = v_{x_{y}}$$

$$K_{2_{r_x}} = v_{x_{i}} + \frac{\tau}{2}.K1_{v_{x}}$$

$$K_{2_{r_y}} = v_{y_{i}} + \frac{\tau}{2}.K1_{v_{y}}$$

$$K_{3_{r_x}} = v_{x_{i}} + \frac{\tau}{2}.K2_{v_{x}}$$

$$K_{3_{r_y}} = v_{y_{i}} + \frac{\tau}{2}.K2_{v_{y}}$$

$$K_{4_{r_x}} = v_{x_{i}} + \tau.K3_{v_{x}}$$

$$K_{4_{r_y}} = v_{y_{i}} + \tau.K3_{v_{y}}$$

Due to how the constants are formulated, the order of calculation is as follows:

- $K_{1_{v_x}}$ and $ K_{1_{v_y}}$
- $K_{1_{r_x}}$ and $ K_{1_{r_y}}$
- $K_{2_{v_x}}$ and $ K_{2_{v_y}}$
- $K_{2_{r_x}}$ and $ K_{2_{r_y}}$
- $K_{3_{v_x}}$ and $ K_{3_{v_y}}$
- $K_{3_{r_x}}$ and $ K_{3_{r_y}}$
- $K_{4_{v_x}}$ and $ K_{4_{v_y}}$
- $K_{4_{r_x}}$ and $ K_{4_{r_y}}$

# Answer: Exercise 3.6

The following function orbit_EC is written based on the pseudo-code for the **Euler-Cromer** method.

This method does the following:
- Approximates the values of rx and ry (the radial position of the comet) and vx and vy (the velocity of the comet) using the **Euler-Cromer method**.
- The method stops calculating when the comet has completed **two orbits**. The method tests with each iteration if the comet has completed a orbit by testing if the value of **ry has changed sign (in comparison with the previous iteration)**, and by testing if **rx is positive**.
- Records the kinetic-, potential and total-energy of the comet when it starts an orbit.
- Calculates how much energy is gained or lost by the comet when it starts it's second orbit, in comparison with it's first orbit.
- Calculates the percentage energy difference between **these values**.
- Plots the orbital path of the comet and energy-vs-time graphs of the comet, only when commanded to do so. This is a convenient way of testing the program.

**Note: Calling Python functions within a loop is a notoriously slow process, therefore as little as possible other Python functions are called.**

## Calculate the orbital path of Halley's comet using the Euler-Cromer method

```python
def orbit_EC(r0, v0, m, tau, plot):
    
    '''
    Parameters
    ----------
    r0: Initial radial distance (AU)
    v0: Initial tangential velocity (AU/yr)
    m: Mass of the comet [kg]
    tau: Time-step [yr]
    plot: [Boolean] - Tells whether or not this function should plot the graphs or not.
    '''
    
    # Physical Constants
    # ------------------
    
    GM = 4 * pi ** 2
    
    # Variables
    # ----------
    
    # Initial radial position of the comet
    rx = [r0]
    ry = [0.00]
    r = [math.sqrt(rx[0]**2 + ry[0]**2)]
    
    # Initial velocity of the comet
    vx = [0.00]
    vy = [v0]
    v = [math.sqrt(vx[0]**2 + vy[0]**2)]
    
    # Initial acceleration of the comet
    ax = [(-GM*rx[0]) / (math.sqrt(rx[0]**2 + ry[0]**2) ** 3)]
    ay = [(-GM*ry[0]) / (math.sqrt(rx[0]**2 + ry[0]**2) ** 3)]
    a = [math.sqrt(ax[0]**2 + ay[0]**2)]
    
    # Initial kinetic-, potential- and total- energy of the comet.
    kinetic = [0.5 * m * v[0] ** 2]
    potential = [-GM * m / r[0]]
    energy = [kinetic[0] + potential[0]]
    
    # Initial time
    time = [0]
    
    # Records the index at which the comet starts a new orbit
    orbit_start_indeces = [0]
    
    # Counter for how many times the comet has completed a full orbit
    orbit_number = 0
    
    # Euler-Cromer Method
    # --------------------
    
    i = 0
    
    while True:
        
        # Calculate the acceleration of the comet
        ax = np.append(ax, (-GM*rx[i]) / (math.sqrt(rx[i]**2 + ry[i]**2) ** 3))
        ay = np.append(ay, (-GM*ry[i]) / (math.sqrt(rx[i]**2 + ry[i]**2) ** 3))
        a = np.append(a, math.sqrt(ax[i]**2 + ay[i]**2) )
        
        # Approximate the velocity of the comet using the Euler-Cromer method
        vx = np.append(vx, vx[i] + tau * ax[i+1])
        vy = np.append(vy, vy[i] + tau * ay[i+1])
        v = np.append(v, math.sqrt(vx[i]**2 + vy[i]**2) )
        
        # Approximate the position of the comet using the Euler-Cromer method
        rx = np.append(rx, rx[i] + tau * vx[i+1])
        ry = np.append(ry, ry[i] + tau * vy[i+1])
        r = np.append(r, math.sqrt(rx[i]**2 + ry[i]**2) )
        
        # Calculate the kinetic-, potential-, and total energy of the comet
        kinetic = np.append(kinetic, 0.5 * m * v[i] ** 2)
        potential = np.append(potential, -GM * m / r[i])
        energy = np.append(energy, kinetic[i] + potential[i])
        
        # Update the time array
        time = np.append(time, i*tau)
        
        # Increment i to continue the loop
        i = i + 1
        
        # Test when the comet starts a new revolution
        if (i>1) and (rx[i] > 0) and (ry[i] * ry[i-1] < 0):
            # When starting a new revolution, record the indeces
            orbit_start_indeces = np.append(orbit_start_indeces, i-1)
            # ... and count the revolution
            orbit_number = orbit_number + 1
            # Stop when finishing the 2nd revolution
            if orbit_number == 2:
                break;
                
    # Calculate the initial total energy of the comet when starting a new orbit        
    energy_peaks = energy[orbit_start_indeces]
    
    # Calculate the total energy difference between when the comet started...
    # ... the second orbit in comparison with when it started the first orbit
    energy_difference = energy_peaks[1] - energy_peaks[0]
    energy_difference_percentage = np.abs((energy_peaks[1] - energy_peaks[0]) / (energy_peaks[0]) * 100)
    
    # Plot when instructed to do so
    if plot is True:
        plot_orbit_path(rx, ry, r0)
        plot_orbit_energies(time, kinetic, potential, energy)
    
    # Return the calculated energy difference as a percentage
    return energy_difference_percentage
```

## Find the best time step for a range of initial velocities

Now that a Python function exists that can calculate the energy difference per orbit **using the Euler-Cromer method**, we can define a second Python function that will determine the best time-step or "best tau" value (a time step where about 1% total energy is conserved per orbit) for any given intial velocity.

This function will calculate the energy difference per orbit for an initial velocity and an inital guess for the time-step. If the energy difference per orbit is not about 1%, the function will decrease the time-step by 5% (this can be changed if desired) and recalculate the energy difference per orbit.

The function will continue decreasing the time-step by 5% repeatedly up until an energy difference per orbit about 1% is obtained. The final time-step, we return as the "best tau" value.

```python
def find_best_tau_EC(Aphelion_Halley, v0_Guess, m_Halley, tau_Guess):
    
    # An array in which to store the 'errors' (the energy difference per orbit as a percentage)
    errors = []
    # Condition for the loop to continue
    flag = False
    # Initial counter
    i = 0

    while flag is False:
        # Calculate the energy difference per orbit by calling the orbit_EC function
        error = orbit_EC(Aphelion_Halley, v0_Guess, m_Halley, tau_Guess, False)
        # Append this value to the errors array
        errors = np.append(errors, error)
        
        # If the energy difference per orbit is about 1%, stop.
        if error < 1:
            flag = True
        
        # If not, continue decreasing the value of the guessed time-step.
        if flag is False:
            tau_Guess = 0.95 * tau_Guess
            i = i +1
    
    # Return the best-tau value for the given initial velocity
    return tau_Guess, error
```

Here, the exact same method is applied as was done to complete Exercise 3.6. The only difference is that the 4th-Order Runge-Kutta method is used in stead of the Euler-Cromer method.

Similar as before, a function is written to calculate the orbital path of the comet. The function 'orbit_RK4' is based off the Pseudo-code listed in the introduction of this report.

This method does the following: *(Exactly the same as the orbit_EC function, but only uses the 4th-Order Runge-Kutta method)*.
- Approximates the values of rx and ry (the radial position of the comet) and vx and vy (the velocity of the comet) using the **4th Order Runge-Kutta method**.
- The method stops calculating when the comet has completed **two orbits**. The method tests with each iteration if the comet has completed a orbit by testing if the value of **ry has changed sign (in comparison with the previous iteration)**, and by testing if **rx is positive**.
- Records the kinetic-, potential and total-energy of the comet when it starts an orbit.
- Calculates how much energy is gained or lost by the comet when it starts it's second orbit, in comparison with it's first orbit.
- Calculates the percentage energy difference between **these values**.
- Plots the orbital path of the comet and energy-vs-time graphs of the comet, only when commanded to do so. This is a convenient way of testing the program.

## Calculate the orbital path of Halley's comet using the 4th Order Runge-Kutta method

```python
def orbit_RK4(r0, v0, m, tau, plot):
    
    '''
    r0: [Float] Initial radial distance (AU)
    v0: [Float] Initial tangential velocity (AU/yr)
    m: [Float] Mass of the comet [kg]
    tau: [Float] Time-step [yr]
    plot: [Boolean] - Tells whether or not this function should plot the graphs or not.
    '''
    
    # Physical Constants
    # ------------------
    
    GM = 4 * pi ** 2
    
    # Variables
    # ----------
    
    rx = [r0]
    ry = [0.00]
    
    r = [math.sqrt(rx[0]**2 + ry[0]**2)]
    
    vx = [0.00]
    vy = [v0]
    v = [math.sqrt(vx[0]**2 + vy[0]**2)]
    
    ax = [(-GM*rx[0]) / (math.sqrt(rx[0]**2 + ry[0]**2) ** 3)]
    ay = [(-GM*ry[0]) / (math.sqrt(rx[0]**2 + ry[0]**2) ** 3)]
    a = [math.sqrt(ax[0]**2 + ay[0]**2)]
    
    kinetic = [0.5 * m * v[0] ** 2]
    potential = [-GM * m / r[0]]
    energy = [kinetic[0] + potential[0]]
    
    time = [0]
    
    orbit_start_indeces = [0]
    orbit_number = 0
    
    # 4th Order Runge-Kutta Method
    # -----------------------------
    
    i = 0
    
    while True:
        
        k1vx = (-GM*rx[i]) / (math.sqrt(rx[i]**2 + ry[i]**2) ** 3)
        k1vy = (-GM*ry[i]) / (math.sqrt(rx[i]**2 + ry[i]**2) ** 3)
        k1rx = vx[i]
        k1ry = vy[i]
        
        k2vx = (-GM*(rx[i]+(tau/2)*(k1rx)) / (math.sqrt((rx[i]+(tau/2)*(k1rx))**2 + (ry[i]+(tau/2)*(k1ry))**2) ** 3))
        k2vy = (-GM*(ry[i]+(tau/2)*(k1ry)) / (math.sqrt((rx[i]+(tau/2)*(k1rx))**2 + (ry[i]+(tau/2)*(k1ry))**2) ** 3))
        k2rx = vx[i] + (tau/2)*(k1vx)
        k2ry = vy[i] + (tau/2)*(k1vy)
                
        k3vx = (-GM*(rx[i]+(tau/2)*(k2rx)) / (math.sqrt((rx[i]+(tau/2)*(k2rx))**2 + (ry[i]+(tau/2)*(k2ry))**2) ** 3))
        k3vy = (-GM*(ry[i]+(tau/2)*(k2ry)) / (math.sqrt((rx[i]+(tau/2)*(k2rx))**2 + (ry[i]+(tau/2)*(k2ry))**2) ** 3))
        k3rx = vx[i] + (tau/2)*(k2vx)
        k3ry = vy[i] + (tau/2)*(k2vy)
                
        k4vx = (-GM*(rx[i]+(tau)*(k3rx)) / (math.sqrt((rx[i]+(tau)*(k3rx))**2 + (ry[i]+(tau)*(k3ry))**2) ** 3))
        k4vy = (-GM*(ry[i]+(tau)*(k3ry)) / (math.sqrt((rx[i]+(tau)*(k3rx))**2 + (ry[i]+(tau)*(k3ry))**2) ** 3))
        k4rx = vx[i] + (tau)*(k3vx)
        k4ry = vy[i] + (tau)*(k3vy)
        
        vx = np.append(vx, vx[i] + (tau/6)*(k1vx + 2*k2vx+ 2*k3vx + k4vx))
        vy = np.append(vy, vy[i] + (tau/6)*(k1vy + 2*k2vy + 2*k3vy + k4vy))
        v = np.append(v, math.sqrt(vx[i]**2 + vy[i]**2) )
        
        rx = np.append(rx, rx[i] + (tau/6)*(k1rx + 2*k2rx + 2*k3rx + k4rx))
        ry = np.append(ry, ry[i] + (tau/6)*(k1ry + 2*k2ry + 2*k3ry + k4ry))
        r = np.append(r, math.sqrt(rx[i]**2 + ry[i]**2) )
        
        kinetic = np.append(kinetic, 0.5 * m * v[i] ** 2)
        potential = np.append(potential, -GM * m / r[i])
        energy = np.append(energy, kinetic[i] + potential[i])
        
        time = np.append(time, i*tau)
        
        i = i + 1
    
        if (i>1) and (rx[i] > 0) and (ry[i] * ry[i-1] < 0):
            orbit_start_indeces = np.append(orbit_start_indeces, i-1)
            orbit_number = orbit_number + 1
            if orbit_number == 2:
                break;
    
    energy_peaks = energy[orbit_start_indeces]
    energy_difference = energy_peaks[1] - energy_peaks[0]
    energy_difference_percentage = abs((energy_peaks[1] - energy_peaks[0]) / (energy_peaks[0]) * 100)
    
    if plot is True:
        plot_orbit_path(rx, ry, r0)
        plot_orbit_energies(time, kinetic, potential, energy)
    
    return energy_difference_percentage
```
